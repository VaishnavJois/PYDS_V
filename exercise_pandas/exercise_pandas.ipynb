```python
allteams = pd.DataFrame()
filelist = ['csk.json','dc.json','kkr.json','kxip.json','mi.json','rcb.json','rr.json','srh.json']
for file in filelist:
    data = pd.read_json(file)
    allteams = allteams.append(data)

allteams
```

```python
allteams.info()
```

```python
allteams.max()
```

```python
price = allteams['Price']
price
```

```python
result = []
for p in price:
    result.append([i for i in p.split()])
    
result
money  = []
for i in result:
    if i[1]=='crore':
        amt = float(i[0])*1000000
    else:
        amt = float(i[0])*100000
    money.append(amt)
money
```

```python
allteams['amt']=money
```

```python
china_corona = pd.DataFrame(corona[corona['Country']=='China'])
china_corona
```

```python
province = china_corona['Province/State'].value_counts()
number_of_province = len(province)
number_of_province
```

```python
corona_confirmed_province = pd.DataFrame(china_corona.groupby('Province/State')['Confirmed'].sum())
corona_confirmed_province
```

```python
corona_confirmed_province['Province'] = corona_confirmed_province.index
corona_confirmed_province
```

```python
corona_C_P = corona_confirmed_province
corona_C_P.index = np.arange(1,number_of_province+1)
```

```python
china_cases = corona_C_P[['Province','Confirmed']]
```

```python
%matplotlib inline

china_cases.plot.bar(x="Province",y="Confirmed")
```
